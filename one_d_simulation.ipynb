```python
import numpy as np
import matplotlib.pyplot as plt
from random import random
```

```python
M = n * n * 5
n = 256
a = 3
x = np.zeros(M)

for i in range(1, M): 
    if random() < .5 * ( 1.0 - a * x[i-1] / n): 
        x[i] = x[i-1] + 1.0 / n
    else:
        x[i] = x[i-1] - 1.0/n

plt.plot(range(M), x, label='O.U.')  # Plot some data on the (implicit) axes.
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title("Discrete O.U.")
plt.legend()
plt.show()
```

```python
m = np.zeros(M)
for i in range(1, M): 
    if x[i] - x[i - 1] > 0:
        m[i] = m[i-1] + np.exp(a * x[i-1] / n + a * (i - M) * 1.0 / (n*n)) / n
    else:
        m[i] = m[i-1] - np.exp(-a * x[i-1] / n + a * (i - M) * 1.0 / (n*n)) / n
                
```

```python
plt.plot(range(M), x, label='O.U.')  # Plot some data on the (implicit) axes.
plt.plot(range(M), m, label='martingale')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title("Discrete O.U.")
plt.legend()
plt.show()
```
